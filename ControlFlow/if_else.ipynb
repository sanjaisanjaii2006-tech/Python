```python
'''
Types:
    1. if
    2. if else
    3. if elif else
    4. nested if

'''
```

```python
'''
1. if
    IF condition:
        statement
'''
name = "sanjai"

if name == "sanjaii":
    print("Verified")
```

```python
'''
2. if
    IF condition:
        statement
    ELSE:
        statement
'''
name = "sanjai"

if name == "sanjaii":
    print("Verified")
else:
    print('Not Verified')
```

```python
'''
3. if
    IF condition:
        statement
    ELIF contition:
        statement
    ELSE:
        statement
'''

age = -2

if age >= 18:
    print("Voter")
elif age < 18 and age > 0:
    print("Not voter")
else:
    print("no aged")
```

```python
# Nested if
no = 34
if no < 20:
    if no > 15:
        print(no)
    else:
        print("no")
else:
    print("main no")
```
